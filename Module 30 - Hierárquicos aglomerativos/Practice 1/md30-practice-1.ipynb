# Mtodos hierrquicos de agrupamento - Tarefa

```python
import pandas              as pd
import matplotlib.pyplot   as plt
import seaborn             as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster       import AgglomerativeClustering

from scipy.cluster         import hierarchy
```

---

## 1) Repetindo o que fizemos em aula

Primeiramente vamos repetir o que fizemos em aula com a base de pinguins.

1. Selecione as variveis quantitativas da base de pinguins
2. Elimine valores faltantes
3. Padronize as variveis para mdia zero e desvio padro 1
4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos
5. Faa o Dendrograma

```python
# 1. Selecione as variveis quantitativas da base de pinguins

# Carregar um conjunto de dados de pinguins (penguins) em um DataFrame chamado 'df'.
df = sns.load_dataset('penguins')

# Filtrar o DataFrame 'df' para incluir apenas colunas que contm tipos de dados numricos.
df_num = df.select_dtypes(include='number')

# Imprimir informaes sobre o DataFrame 'df_num', incluindo o nmero de colunas, o nmero de valores no nulos e os tipos de dados de cada coluna.
print(df_num.info())

# Exibir o DataFrame 'df_num', que agora contm apenas colunas com tipos de dados numricos.
df_num
```

```python
# 2. Elimine valores faltantes

# Exibir as linhas no DataFrame 'df_num' que contm valores faltantes (NaN).
display(df_num[df_num.isna().any(axis=1)])

# Remover as linhas com valores faltantes do DataFrame 'df_num' e atualizar o DataFrame com as alteraes.
df_num.dropna(inplace=True)

# Exibir o DataFrame 'df_num' aps a remoo das linhas.
df_num
```

```python
# 3. Padronize as variveis para mdia zero e desvio padro 1

# Criar uma instncia do objeto StandardScaler, que ser usado para padronizar os dados.
scaler = StandardScaler()

# Aplicar o mtodo fit_transform() do StandardScaler para padronizar os dados numricos no DataFrame 'df_num'.
df_num_pad = scaler.fit_transform(df_num)

# Criar um novo DataFrame a partir dos dados padronizados e exibi-lo.
pd.DataFrame(df_num_pad)
```

```python
# 4. Faa um agrupamento hierrquico dividindo essa base em 3 grupos

# Criar uma instncia do modelo de clustering AgglomerativeClustering com os seguintes parmetros:
    # - n_clusters: O nmero de clusters desejado, neste caso, definido como 3.
    # - linkage: O mtodo de ligao utilizado para calcular a distncia entre clusters. No caso, 'complete' indica ligao completa.
    # - distance_threshold: Define um limite de distncia opcional, que, quando atingido, resulta na parada da aglomerao.
clustering = AgglomerativeClustering(n_clusters=3, 
                                     linkage='complete', 
                                     distance_threshold=None)

# Aplicar o modelo de clustering aos dados padronizados no DataFrame 'df_num_pad'.
clustering.fit(df_num_pad)
```

```python
# Adicionar uma nova coluna chamada 'grupo' ao DataFrame 'df_num' para armazenar os rtulos de cluster gerados pelo modelo de clustering.
df_num['grupo'] = clustering.labels_

# Exibir o DataFrame 'df_num' com a nova coluna 'grupo' que contm as etiquetas de cluster atribudas a cada instncia de dados.
df_num
```

```python
# 5. Faa o Dendrograma

# Definir o tamanho da figura para a plotagem do dendrograma hierrquico.
plt.figure(figsize=(24, 6))

# Calcular a matriz de ligao (linkage matrix) para o dendrograma hierrquico usando os dados padronizados em 'df_num_pad'.
Z = hierarchy.linkage(df_num_pad, method='complete')

# Plotar o dendrograma hierrquico com as informaes contidas em 'Z'.
dn = hierarchy.dendrogram(Z)

# Definir o ttulo do dendrograma.
plt.title('Dendrograma Hierrquico')

# Remover os marcadores no eixo x (rtulos das instncias) para melhorar a legibilidade do dendrograma.
plt.xticks([])

# Definir o rtulo do eixo y como 'Distncia'.
plt.ylabel('Distncia')

# Exibir o dendrograma hierrquico.
plt.show()
```

## 2) Avalie os grupos

Descreva os grupos:

- Quantos pinguins de cada espcie h em cada grupo?
- Descreva as variveis quantitativas de cada grupo com as tcnicas que voc preferir.

```python
print('Quantos pinguins de cada espcie h em cada grupo?\n')

# Contar o nmero de pinguins em cada grupo e classificar os resultados por ndice (rtulo do grupo).
contagem_grupos = df_num['grupo'].value_counts().sort_index()

# Usar um loop para imprimir o nmero de pinguins em cada grupo.
for indice, valor in enumerate(contagem_grupos):
    print('Grupo {}: {} pinguins'.format(indice, valor))

# Criar rtulos para os grupos com base nos ndices dos grupos.
rotulos_grupos = [f'Grupo {grupo}' for grupo in contagem_grupos.index]

# Criar um grfico de barras com os rtulos dos grupos e a contagem de pinguins em cada grupo.
plt.bar(rotulos_grupos, contagem_grupos)

# Adicionar rtulo ao eixo y e um ttulo ao grfico.
plt.ylabel('Nmero de Pinguins')
plt.title('Contagem de Pinguins por Grupo')

# Exibir o grfico de barras.
plt.show()
```

```python
# Criar uma tabela cruzada (crosstab) entre as colunas 'species' e 'grupo' nos DataFrames 'df' e 'df_num'.
tabela_cruzada = pd.crosstab(df['species'], df_num['grupo'])

# Plotar a tabela cruzada como um grfico de barras empilhadas para visualizar a distribuio de espcies em cada grupo.
ax = tabela_cruzada.plot(kind='bar', stacked=True)

# Ajustar a rotao dos rtulos no eixo x para melhorar a legibilidade.
plt.xticks(rotation=45)

# Definir o rtulo do eixo y como 'Nmero de Pinguins' e dar um ttulo ao grfico.
plt.ylabel('Nmero de Pinguins')
plt.title('Distribuio de Espcies em Cada Grupo')

# Adicionar uma legenda ao grfico para identificar os grupos.
plt.legend(labels=[f'Grupo {grupo}' for grupo in tabela_cruzada.columns])

# Adicionar os nmeros inteiros dentro de cada barra (exceto se o nmero for 0).
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if height != 0:
        ax.annotate(int(height), (x + width/2, y + height/2), ha='center', va='center')

# Exibir a tabela cruzada como uma tabela no formato tabular.
display(tabela_cruzada)

# Exibir o grfico de barras empilhadas com a quantidade dentro de cada grupo.
plt.show()
```

```python
# Usar a biblioteca Seaborn para criar um grfico de matriz de disperso (pair plot) com os dados em 'df_num'.
# A opo 'hue'  usada para colorir os pontos com base na coluna 'grupo'.
# A opo 'palette' define a paleta de cores a ser usada para os grupos.
sns.pairplot(data=df_num, hue='grupo', palette='colorblind');
```

## 3) Alterando o agrupamento

O Dendrograma sugere que podem haver grupos interessantes se solicitarmos mais que 3 grupos. Se quisermos testar o agrupamento com mais de 3 grupos, quantos grupos o Dendrograma te sugere? (Dica: no faa mais que 6 grupos).

- Faa o agrupamento hierrquico semelhante ao anterior, mas com o nmero de grupos que voc escolheu.
- Analise esses agrupamentos
    - Quantos pinguins de cada espcie h em cada grupo?
    - Quantos pinguins de cada espcie machos e fmeas h em cada grupo?
- Na sua opinio, o agrupamento capturou algum padro natural dos dados? Qual?

```python
# Configurar o tamanho da figura para o dendrograma.
plt.figure(figsize=(24, 6))

# Calcular a ligao hierrquica usando o mtodo de ligao 'complete' nos dados padronizados em 'df_num_pad'.
Z = hierarchy.linkage(df_num_pad, method='complete')

# Plotar o dendrograma usando os resultados da ligao hierrquica.
# A opo 'color_threshold' define um limite de cor para destacar os clusters.
dn = hierarchy.dendrogram(Z, color_threshold=3.8)

# Remover os rtulos do eixo x para melhorar a legibilidade.
plt.xticks([])

# Exibir o dendrograma hierrquico.
plt.show()
```

```python
# Criar uma instncia do modelo de clustering AgglomerativeClustering com 6 clusters,
# mtodo de ligao 'complete' e sem limite de distncia.
clustering = AgglomerativeClustering(n_clusters=5, linkage='complete', distance_threshold=None)

# Aplicar o modelo de clustering aos dados padronizados no DataFrame 'df_num_pad'.
clustering.fit(df_num_pad)

# Adicionar uma nova coluna chamada 'subgrupo' ao DataFrame 'df_num' para armazenar as etiquetas de subgrupo geradas pelo modelo de clustering.
df_num['subgrupo'] = clustering.labels_

# Exibir o DataFrame 'df_num' com a nova coluna 'subgrupo' que contm as etiquetas de subgrupo atribudas a cada instncia de dados.
df_num
```

```python
print('Quantos pinguins de cada espcie h em cada grupo?\n')

# Contar o nmero de pinguins em cada subgrupo e classificar os resultados por ndice (rtulo do subgrupo).
contagem_subgrupos = df_num['subgrupo'].value_counts().sort_index()

# Usar um loop para imprimir o nmero de pinguins em cada subgrupo.
for indice, valor in enumerate(contagem_subgrupos):
    print('Subgrupo {}: {} pinguins'.format(indice, valor))

# Criar rtulos para os subgrupos com base nos ndices dos subgrupos.
rotulos_subgrupos = [f'Subgrupo {grupo}' for grupo in contagem_subgrupos.index]

# Criar um grfico de barras com os rtulos dos subgrupos e a contagem de pinguins em cada subgrupo.
plt.bar(rotulos_subgrupos, contagem_subgrupos)

# Adicionar um ttulo ao grfico, rtulo ao eixo y e ajustar o tamanho dos rtulos no eixo x.
plt.title('Contagem de Pinguins por Subgrupo')
plt.ylabel('Nmero de Pinguins')
plt.xticks(fontsize=8)

# Adicionar as quantidades dentro de cada coluna do grfico de barras.
for i, v in enumerate(contagem_subgrupos):
    plt.text(i, v, str(v), ha='center', va='bottom')

# Exibir o grfico de barras.
plt.show()
```

```python
# Mesclar o DataFrame 'df_num' com as colunas 'sex' e 'species' do DataFrame original 'df' usando o ndice como chave de mesclagem.
df_ = df_num.merge(right=df[['sex', 'species']], how='left', left_index=True, right_index=True)

# Contar a frequncia dos pares de valores nas colunas 'subgrupo' e 'sex' usando a funo value_counts.
# Em seguida, desempilhar o resultado para criar uma tabela cruzada.
# Qualquer valor ausente (NaN)  preenchido com zero (0) e os valores so convertidos para inteiros.
contagem_subgrupo_sex = (df_[['subgrupo', 'sex']].value_counts().unstack().fillna(0).astype(int))

print('Quantos pinguins de cada espcie machos e fmeas h em cada grupo?')

# Exibir a tabela cruzada com a contagem de pinguins por subgrupo e sexo.
display(contagem_subgrupo_sex)

print('\nTabela cruzada percentual:')

# Calcular a tabela cruzada percentual das espcies por subgrupo e sexo.
tabela_cruzada_percentual = pd.crosstab(index=[df_['subgrupo'], df_['sex']], columns=df_['species'], normalize=True) * 100
tabela_cruzada_percentual = tabela_cruzada_percentual.unstack().fillna(0)

# Exibir a tabela cruzada percentual com valores formatados para duas casas decimais.
display(tabela_cruzada_percentual.applymap(lambda x: f'{x:.2f}%'))

# Plotar a tabela cruzada como um grfico de barras.
contagem_subgrupo_sex.plot(kind='bar')

# Adicionar um ttulo ao grfico, definir a rotao dos rtulos no eixo x e exibir o grfico de barras.
plt.title('Contagem de Pinguins por Subgrupo e Sexo')
plt.xticks(rotation=0)
plt.show()
```

```python
# Criar um pair plot (grfico de pares) usando o DataFrame 'df_'.
# A opo 'hue'  usada para colorir os pontos com base no novo agrupamento.
# A opo 'palette' define a paleta de cores a ser usada.
sns.pairplot(data=df_, hue='subgrupo', palette='colorblind');
```
