A ausncia de padronizao pode ocasionar um desequilbrio na influncia exercida pelas variveis na distncia entre os pontos, dependendo da escala empregada. Isso pode comprometer a identificao equitativa de agrupamentos no futuro, visto que variveis com escalas mais amplas podem predominar no processo de clusterizao em detrimento daquelas com escalas menores. Portanto, antes de prosseguir com o processo de clusterizao,  aconselhvel aplicar a padronizao (normalizao) das variveis para assegurar que todas contribuam de forma igual para a formao dos agrupamentos. Tal procedimento viabilizar uma anlise mais justa e precisa dos padres subjacentes aos dados, resultando em uma clusterizao mais confivel e significativa.

### Preparao dos dados (Padronizao):

```python
# Cria uma cpia do DataFrame df.
df_pad = df.copy()

# Seleciona somente as colunas numricas do DataFrame df_pad.
colunas_numericas = df_pad.select_dtypes(include = 'number').columns

# Aplica a padronizao nas colunas numricas do DataFrame df_pad, substituindo os valores das colunas padronizadas.
# As colunas so padronizadas usando a classe StandardScaler do scikit-learn.
# A classe StandardScaler centraliza os dados para ter mdia zero e escalona-os para ter desvio padro igual a um.
df_pad[colunas_numericas] = pd.DataFrame(data=StandardScaler().fit_transform(df_pad[colunas_numericas]), 
                                         columns=df_pad[colunas_numericas].columns)

# O DataFrame df_pad agora contm as colunas numricas padronizadas.
df_pad
```

## Nmero de grupos

Utilize as tcnicas vistas em aula que te ajudem a decidir pelo nmero de grupos a ser utilizados.

### Mtodos para analisar a quantidade de clusters
#### Mtodo do Cotovelo

```python
# Inicializao de uma lista vazia para armazenar as somas dos quadrados das distncias (SQD).
SQD = []

# Nmero mximo de clusters.
K = 10

# Loop para iterar sobre diferentes valores de k (nmero de clusters).
# Calcula o K-Means para cada valor de k e armazena a soma dos quadrados das distncias na lista SQD.
for k in tqdm(range(1, K + 1)):
    km = KMeans(n_clusters = k, 
                n_init = 'auto')
    km = km.fit(df_pad[colunas])
    SQD.append(km.inertia_)

# Criao de um DataFrame com os valores de k e suas respectivas somas dos quadrados das distncias (SQD).
df_sqd = pd.DataFrame(data = {'n_clusters': list(range(1, len(SQD) + 1)), 
                              'SQD': SQD})
# Exibe o DataFrame df_sqd com os valores de SQD para cada nmero de clusters.
print(df_sqd)

# Criao do grfico para visualizao do mtodo do cotovelo.
plt.plot(df_sqd['n_clusters'], 
         df_sqd['SQD'])
plt.xlabel('Nmero de Clusters (k)')
plt.ylabel('Soma dos Quadrados das Distncias (SQD)')
plt.title('Mtodo do Cotovelo para Determinar o Nmero de Clusters')
plt.show()
```

#### Coeficiente da Silhueta

```python
# Inicializao de uma lista vazia para armazenar os valores do Coeficiente da Silhueta.
silhuetas = []

# Nmero mximo de clusters a serem considerados (de 2 a max_clusters).
max_clusters = 10

# Loop para iterar sobre diferentes nmeros de clusters.
# Calcula o K-Means para cada valor de n_clusters e armazena o Coeficiente da Silhueta na lista 'silhuetas'.
for n_clusters in tqdm(range(2, max_clusters + 1)):
    
    # Executa o K-Means com o nmero de clusters atual.
    km = KMeans(n_clusters = n_clusters, 
                n_init = 'auto').fit(df_pad[colunas])
    
    # Calcula o Coeficiente da Silhueta e adiciona  lista 'silhuetas'.
    silhuetas.append(silhouette_score(X = df_pad[colunas], 
                                      labels = km.labels_))

    # Criao de nomes para os grupos identificados pelo K-Means.
    nomes_grupos = ['grupo_' + str(n) for n in range(n_clusters)]
    
    # Adiciona uma nova coluna ao DataFrame df_pad_ com os rtulos dos grupos.
    df_pad['grupos_'+str(n_clusters)] = pd.Categorical.from_codes(codes = km.labels_, 
                                                                  categories = nomes_grupos)    

# Criao de um DataFrame com os valores do Coeficiente da Silhueta mdio para cada nmero de clusters.
df_silhueta = pd.DataFrame(data = {'n_clusters': list(range(2, max_clusters + 1)), 
                                   'silhueta_media': silhuetas})
# Exibe o DataFrame com os valores do Coeficiente de Silhueta mdio para cada nmero de clusters.
print(df_silhueta)

# Criao do grfico de linha para visualizao do Coeficiente da Silhueta mdio em relao ao nmero de clusters.
df_silhueta.plot.line(x = 'n_clusters', 
                      y = 'silhueta_media', 
                      marker = 'o')

plt.show()
```

## Avaliao dos grupos

Faa uma anlise descritiva para pelo menos duas solues de agrupamentos (duas quantidades diferentes de grupos) sugeridas no item anterior, utilizando as variveis que esto no escopo do agrupamento.
- Com base nesta anlise e nas anlises anteriores, decida pelo agrupamento final. 
- Se puder, sugira nomes para os grupos.

```python
def cluster_barplot(data: pd.DataFrame, coluna_cluster: str) -> None:
    """
    Cria um grfico de barras para mostrar a quantidade de clientes por cluster.

    Parmetros:
        data (pd.DataFrame): O DataFrame contendo os dados.
        coluna_cluster (str): O nome da coluna que contm as informaes dos clusters.

    Retorna:
        None
    """
    # Agrupa os dados pelo cluster e por 'Revenue', contando a quantidade de clientes em cada combinao.
    df_agrupado = data.groupby([coluna_cluster, 'Revenue'])['Revenue'].count().unstack()
    
    # Cria um grfico de barras a partir do DataFrame agrupado.
    ax = df_agrupado.plot.bar(rot=0)
    
    # Define rtulos para os eixos X e Y, alm do ttulo do grfico.
    ax.set_ylabel('Quantidade de clientes')
    ax.set_xlabel('Clusters')
    plt.title('Quantidade de clientes por cluster')
    
    # Exibe o grfico.
    plt.show()

# Exemplo de uso:

# Cria um grfico de barras para mostrar a quantidade de clientes por cluster (2 clusters).
cluster_barplot(data = df_pad, coluna_cluster = 'grupos_2')

# Cria um grfico de barras para mostrar a quantidade de clientes por cluster (3 clusters).
cluster_barplot(df_pad, 'grupos_3')

# Cria um grfico de barras para mostrar a quantidade de clientes por cluster (4 clusters).
cluster_barplot(df_pad, 'grupos_4')
```

```python
# Cria uma cpia do DataFrame padronizado df_pad.
df_pad_ = df_pad.copy()

# Renomeia as duas categorias na coluna 'grupos_2' para rtulos mais descritivos.
df_pad_['grupos_2'] = df_pad_['grupos_2'].cat.rename_categories({'grupo_0': 'No interessados', 
                                                                 'grupo_1': 'Compradores'})

# Renomeia as trs categorias na coluna 'grupos_3' para rtulos mais descritivos.
df_pad_['grupos_3'] = df_pad_['grupos_3'].cat.rename_categories({'grupo_0': 'No interessados', 
                                                                 'grupo_1': 'Potenciais compradores', 
                                                                 'grupo_2': 'Compradores'})
```

```python
# Cria um grfico de barras para mostrar a quantidade de clientes por cluster (2 clusters) usando as categorias renomeadas.
cluster_barplot(df_pad_, 'grupos_2')

# Cria um grfico de barras para mostrar a quantidade de clientes por cluster (3 clusters) usando as categorias renomeadas.
cluster_barplot(df_pad_, 'grupos_3')
```

Apesar da disparidade na distribuio dos clientes classificados como False e True na varivel 'Revenue', o grupo_0 conseguiu identificar a maioria dos clientes, os quais demonstram menor propenso  compra. Surpreendentemente, este grupo acertou aproximadamente 74% dos casos na base de dados onde a varivel 'Revenue'  False (indicando no-compradores), abrangendo 85% dos registros totais. Por outro lado, o grupo_1, encarregado de identificar os clientes mais inclinados a comprar, acertou aproximadamente 4% dos casos onde a varivel 'Revenue'  True (indicando compradores), representando 15% dos dados totais da base.
